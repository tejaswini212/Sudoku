```python
#importing the necessary libraries
import tensorflow as tf
from tensorflow.keras.models import Sequential 
from tensorflow.keras.layers import Dense,Dropout,Activation,Flatten,Conv2D,MaxPooling2D
import pickle
import numpy as np
from sklearn.model_selection import train_test_split
from matplotlib import pyplot as plt
```

```python
#Using the dumps of data we've created earlier
X = pickle.load(open("X.pickle","rb"))
Y = pickle.load(open("Y.pickle","rb"))

Y = np.array(Y) #Converting the Y list of values to numpy array
X = X/255.0 #Normalize our data
```

### Let's get moving to the ML part our task

```python
#Defining a function to train model based on customised parameter input
def custom_train_model(X,Y,ip_test_size=0.3,ip_random_state=7,ip_epochs=10):
    #Important step - Splitting the data!!
    X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=ip_test_size,random_state=ip_random_state)

    #Deciding upon the architecture of our model
    model = tf.keras.models.Sequential([
      tf.keras.layers.Conv2D(64, (3,3), activation='relu', input_shape= X.shape[1:]),
      tf.keras.layers.MaxPooling2D(2, 2),
      tf.keras.layers.Conv2D(64, (3,3), activation='relu'),
      tf.keras.layers.MaxPooling2D(2,2),
      tf.keras.layers.Flatten(),
      tf.keras.layers.Dense(128, activation='relu'),
      tf.keras.layers.Dense(10, activation='softmax')
    ])

    model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

    #Getting to know more about the layers of our architecture
    model.summary()

    #Training the data
    model.fit(X_train, Y_train, epochs=ip_epochs)

    #Evaluating the model:
    test_loss,test_acc = model.evaluate(X_test, Y_test)
    
    return test_loss,test_acc    
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.3,random_state=7)

#Deciding upon the architecture of our model
model = tf.keras.models.Sequential([
  tf.keras.layers.Conv2D(64, (3,3), activation='relu', input_shape= X.shape[1:]),
  tf.keras.layers.MaxPooling2D(2, 2),
  tf.keras.layers.Conv2D(64, (3,3), activation='relu'),
  tf.keras.layers.MaxPooling2D(2,2),
  tf.keras.layers.Flatten(),
  tf.keras.layers.Dense(128, activation='relu'),
  tf.keras.layers.Dense(10, activation='softmax') #for binary classifier sigmoid can be used
])

model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

#Getting to know more about the layers of our architecture
model.summary()

#Training the data
model.fit(X_train, Y_train, epochs=5)

#Evaluating the model:
test_loss,test_acc = model.evaluate(X_test, Y_test)

print("Accuracy: ",test_acc,"Loss: ",test_loss)
predictions = model.predict([X_test]) # Make prediction
```
